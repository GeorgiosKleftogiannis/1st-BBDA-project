# 1st-BBDA-project

This is an EDA project that uses the Used Cars Price Data provided by [Kaggle](https://www.kaggle.com/datasets/avikasliwal/used-cars-price-prediction/data?select=train-data.csv).

## 1. Using git

   <u>new branch (Assumes that repo already exists)<u>
1. git checkout dev ➡ get in dev branch
3. git pull ➡ updates the dev
5. git checkout -b "branch_name" ➡ creates and move into a new branch
6. git merge dev ➡ gets the last updates of dev branch
7. git push --set -upstream origin "branch_name" ➡ First push for the new branch
8. PR for commit changes

<u> extra commands (Assumes that repo already exists)<u>
- git branch -d "branch_name" ➡ safely delete the local branch
