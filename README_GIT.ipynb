## Using git

   <u>new branch (Assumes that repo already exists)<u>
1. git checkout dev ➡ get in dev branch
2. git pull ➡ updates the dev
3. git checkout -b "branch_name" ➡ creates and move into a new branch
4. git merge dev ➡ gets the last updates of dev branch
5. git push --set -upstream origin "branch_name" ➡ First push for the new branch

    *or* git push
6. PR for commit changes

<u> PR for commit changes<u>
1. Go to GitHub page
2. Select the repo
3. Pull requests
4. New Pull request
5. Choose from the list your branch
6. Check the changes
7. Create pull request
8. Add a description (optional)
9. Create pull request

<u> Reviewer role <u>
1. Check the changes
   
   ➡ accept *or*
   
   ➡ send message with corrections
   
3. If changes are accepted, merge
4. Confirm merge
5. Delete branch

<u> extra commands (Assumes that repo already exists)<u>
- git branch -d "branch_name" ➡ safely delete the local branch
